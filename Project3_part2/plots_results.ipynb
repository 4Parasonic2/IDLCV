```python
import json
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import argparse
import random
import glob
```

```python


json_files = sorted(glob.glob("results_arrays/results_p*_n*.json"))
labels = [Path(f).stem.replace("results_", "") for f in json_files]
labels = ['20 clicks', '50 clicks', '5 clicks']

# Load all
experiments = []
for f in json_files:
    with open(f) as jf:
        experiments.append(json.load(jf))

# plt.style.use('seaborn-v0_8')
plt.style.use('default')        # clean base
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['figure.facecolor'] = 'white'
colors = plt.cm.tab10(np.linspace(0, 1, len(experiments)))
colors = {'#3498db', 
          '#e74c3c', 
          '#2ecc71'}
# 'weak_clicks_pos20_neg20'

metrics = ["train_loss", "val_dice", "val_iou", "val_acc", "val_sens", "val_spec"]
titles = {
    "train_loss": "Training Loss",
    "val_dice": "Validation Dice",
    "val_iou": "Validation IoU",
    "val_acc": "Validation Accuracy",
    "val_sens": "Validation Sensitivity",
    "val_spec": "Validation Specificity"
}

# for metric in metrics:
#     plt.figure(figsize=(10, 6))
#     for data, label, color in zip(experiments, labels, colors):
#         if metric not in data:
#             continue
#         y = np.array(data[metric])
#         x = np.arange(len(y))
#         plt.plot(x, y, label=label, color=color, linewidth=2, marker='o', markersize=3)

#     plt.title(titles[metric], fontsize=16, fontweight='bold')
#     plt.xlabel("Epoch")
#     plt.ylabel(metric.replace("_", " ").title())
#     plt.grid(True, alpha=0.3)
#     plt.legend(title="Experiment")
#     plt.tight_layout()

#     out_path = Path("./") / f"{metric}.png"
#     plt.savefig(out_path, dpi=200, bbox_inches='tight')
#     plt.savefig(out_path.with_suffix(".pdf"), bbox_inches='tight')
#     plt.close()

# fig, axes = plt.subplots(2, 4, figsize=(20, 10), sharex=True)
fig, axes = plt.subplots(3, 3, figsize=(18, 15), sharex=True)
axes = axes.flatten()  # [0..7]



# 6. Plot each metric
for idx, metric in enumerate(metrics):
    ax = axes[idx]
    for data, label, color in zip(experiments, labels, colors):
        if metric not in data:
            continue
        y = np.array(data[metric])
        x = np.arange(len(y))
        ax.plot(x, y, label=label, color=color, linewidth=2, marker='o', markersize=3)

    ax.set_title(titles[metric], fontsize=14, fontweight='bold')
    ax.set_ylabel(metric.replace("_", " ").replace("val ", "Val "))
    ax.grid(True, alpha=0.3)
    # if idx >= 4:  # bottom row
    ax.set_xlabel("Epoch")
    
    ax.legend(title="Experiment", fontsize=9, title_fontsize=10, loc="upper right")

# 7. Hide unused subplots (2 empty: index 6 and 7)
# for idx in range(len(metrics), 8):
#     fig.delaxes(axes[idx])

for idx in range(len(metrics), 9):
    fig.delaxes(axes[idx])

# 8. Final layout
plt.suptitle("Weak Annotation Results", fontsize=18, fontweight='bold', y=0.98)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# 9. Save ONLY ONE PNG
out_path = Path("./") / "part2_metrics.png"
plt.savefig(out_path, dpi=300, bbox_inches='tight')
plt.close()

print(f"Plotted: {out_path}")
```
